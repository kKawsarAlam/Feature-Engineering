**Label Encoding**   

Label encoding is a data preprocessing technique in machine learning that converts categorical text data into numerical integers, assigning a unique number (e.g., 0, 1, 2) to each category. It enables algorithms to process non-numeric data, such as converting "Red", "Blue", "Green" into \(0,1,2\), but assumes an ordered relationship between categories, which may not always exist.  

**Key Aspects of Label Encoding:**   

Purpose: Converts categorical labels into a numeric form for model input.   
Methodology: Assigns a unique integer to each unique category in a dataset.  
Use Case: Often used for target variables (y) in classification problems.   
Limitation: It implies an ordinal relationship (e.g.,2>1>0), which can mislead algorithms if the categories are nominal (e.g., country names, colors).  
Example: In a feature with ['Low', 'Medium', 'High'], label encoding might assign (0,1,2) to represent the order.

```python
# Import Libraries
import pandas as pd
import numpy as np

print("Successfully Libraries Imported")
```

```python
# Load Dataset
df = pd.read_csv('/content/home_prices_simple.csv')
df
```

**Label Encoding**  

**Method 1: Using map()**

```python
label_map = {
    'monroe township': 0,
    'west windsor': 1,
    'robbinsville': 2
}

df['town_encoded'] = df['town'].map(label_map)
df
```

**Method 2: Using replace()**

```python
df['town_encoded'] = df['town'].replace({
    'monroe township': 0,
    'west windsor': 1,
    'robbinsville': 2
})
df
```

**Method 3: Auto-create labels**

```python
# Get unique categories
categories = df['town'].unique()
```

```python
# Create mapping automatically
label_map = {cat: idx for idx, cat in enumerate(categories)}
```

```python
# Apply encoding
df['town_encoded'] = df['town'].map(label_map)
df
```

**Using Sklearn LabelEncoder()**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['town_encoded'] = le.fit_transform(df['town'])
df
```

```python
# Drop some redundant columns so that it can not mix up
df = df.drop(columns=['town'])
df
```

**Model Traning**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
Y = df.price
Y
```

```python
model.fit(X, Y)
```

```python
# Predict the price for monroe township [0]
print("monroe township:", model.predict([[3400, 0]]))
```

```python
# Predict the price for west windsor [0]
print("west windsor:", model.predict([[2800, 2]]))
```

```python
# Predict the price for monroe township [0, 0]
print("robbinsville:", model.predict([[3000, 1]]))
```

```python
# Model Accuracy
score = model.score(X, Y)
print(f"Model Score: {score:.2f}")
```
